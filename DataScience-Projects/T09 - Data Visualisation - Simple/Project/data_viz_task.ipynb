```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
selected_df = df[df['Manufacturer'].isin(selected_manufacturers)]

plt.figure(figsize=(10, 6))
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=selected_df, order=selected_manufacturers)
plt.title('Revs per Mile for Selected Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Manufacturer with the highest revs per mile: Suzuki Highest revs per mile: 10.6

### 2. Histogram

```python
# Histogram of MPG
plt.figure(figsize=(10, 6))
sns.histplot(df['MPG.city'], color='skyblue', label='MPG in City', kde=True)
sns.histplot(df['MPG.highway'], color='salmon', label='MPG on Highway', kde=True)
plt.title('Histogram of MPG in City vs. on Highway')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:** On average, driving on the highway is more fuel-efficient compared to driving in the city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure(figsize=(10, 6))
sns.lineplot(x='Wheelbase', y='Turn.circle', data=df)
plt.title('Relationship between Wheelbase and Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The relationship between wheelbase and turning circle size is essentially how the size of the vehicle's wheelbase affects its turning ability. In general, when the wheelbase gets larger, the turning circle tends to increase as well. This means that vehicles with longer wheelbases typically require more space to make a complete turn compared to vehicles with shorter wheelbases. The illustrates shows this relationship by showing how turning circle size varies with different wheelbase lengths across the dataset.

### 4. Barplot

```python
# Barplot for mean horsepower
plt.figure(figsize=(10, 6))
sns.barplot(x='Type', y='Horsepower', data=df, estimator='mean')
plt.title('Mean Horsepower for Each Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

Yes larger cars tend to have more horsepower
