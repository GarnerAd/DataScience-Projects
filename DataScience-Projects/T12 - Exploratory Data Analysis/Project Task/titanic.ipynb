```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv("Titanic.csv")
```

```python
# Replace missing values (NaN) with 0
titanic_df.fillna(0, inplace=True)

# Display the first few rows of the dataset
titanic_df.head()
```

What is the most important factor in determining survival of the Titanic
incident?

```python
# Let's explore the relationship between survival and passenger class
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Let's visualize survival count by passenger class
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

Women and children first. Was this the case?

```python
# Let's visualize survival count by gender
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize survival count by age category
sns.histplot(x='Age', hue='Survived', data=titanic_df, bins=20, kde=True)
plt.title('Survival Count by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Add one other observation that you have noted in the dataset
# Let's explore the relationship between fare paid and survival
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=titanic_df)
plt.title('Survival by Fare Paid')
plt.xlabel('Survived')
plt.ylabel('Fare Paid')
plt.show()
```
